```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from os import listdir
from sklearn.metrics import accuracy_score,classification_report,classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier ,  RandomForestRegressor
import seaborn as sns
import math
from sklearn.tree import DecisionTreeClassifier
import warnings
warnings.filterwarnings("ignore",category=FutureWarning)
```

```python
df = pd.read_csv('./train.csv')
df.head()
```

```python
test = pd.read_csv('./test.csv')
test.head()
```

```python
# Gender Segregation
```

```python
#Female 
```

```python
women = df[df['Sex'] == 'female']
total_women = len(women)
women = women[women['Survived'] == 1]
women.head()
```

```python
survival_women = round((len(women)/total_women)*100,6)
print(f'% Survival of women = {survival_women}')
```

```python
#Male
```

```python
men = df[df['Sex'] == 'male']
total_men = len(men)
men = men[men['Survived'] == 1]
men.head()
```

```python
survival_men = round((len(men)/total_men)*100,6)
print(f'% Survival of male = {survival_men}')
```

```python
# Applied Machine Learning - Random Forest Classifier
```

```python
y = df['Survived']
```

```python
specs = ['Sex','SibSp','Pclass','Parch']
```

```python
x = pd.get_dummies(df[specs])
```

```python
x_test = pd.get_dummies(test[specs])
```

```python
model = RandomForestClassifier(n_estimators=100,max_depth=5,random_state=1)
model.fit(x,y)
preds = model.predict(x_test)
```

```python
out = pd.DataFrame({'PassengerId' : test.PassengerId , 'Survived' : preds})
```

```python
out.to_csv('my_submission1.csv',index=False)
```

```python
decision_tree = DecisionTreeClassifier()
decision_tree.fit(x, y)
Y_pred = decision_tree.predict(x_test)
acc_decision_tree = round(decision_tree.score(x, y) * 100, 2)
acc_decision_tree
```

```python
out = pd.DataFrame({'PassengerId' : test.PassengerId , 'Survived' : Y_pred})
```

```python
out.to_csv('my_submission.csv',index=False)
```
